### While

Supongamos que queremos calcular la suma de los números de 1 a 100. Obviamente, hacer eso _a mano_ es insalubre. Podemos utilizar la sentencia **while**, cuya sintaxis es:

$$\begin{array}{ll}{\textbf{while}}& \textrm{condicion}:\\
 & \textrm{instrucciones}\end{array}$$

```python
i    = 0 #contador de iteraciones
suma = 0 
while i<=100:
    suma = suma + i
    i    = i+1
print('El valor de la suma es:', suma)
```

### Operadores binarios:

Para escribir condiciones suelen usarse operadores binarios, que realizan comparaciones. Algunos ejemplos son:
$$\begin{array}{ll} \textrm{menor} & <\\
                    \textrm{mayor} & >\\
                    \textrm{menor o igual} & <=\\
                    \textrm{mayor o igual} & >=\\
                    \textrm{igual} & == \\
                    \textrm{distinto} & !=\end{array}$$
También se pueden combinar varias condiciones o negar alguna condición usando:
$$\begin{array}{ll} \textrm{y} & \textbf{and}\\
                    \textrm{o} & \textbf{or}\\
                    \textrm{negar} & \textbf{not}
                    \end{array}$$

Escribamos un programa que calcule la suma de los elementos de la siguiente lista:

```python
a = [1,5,7,10,23,1,6,8,12]
print(len(a))
```

```python
i    = 0
suma = 0
while i<len(a):
    suma = suma + a[i]
    i    = i+1
print('La suma es:',suma)
```

### For

El **for** es una secuencia simplificada que permite hacer operaciones similares a las que hicimos anteriormente, pero sin escribir una condición. Al **for** hay que indicarle la _cantidad de pasos_ que tiene que dar. Por ejemplo, la suma de los números de 1 a 100 se puede programar así:

```python
suma = 0
for i in range(101):
    suma = suma+i
print('La suma es:',suma)
```

El comando $range$ va generando automáticamente los números desde 0 hasta 101 (sin contar el último). En el **for** no hay que definir previamente el índice ($i$), ni actualizar su valor. Por lo tanto, los siguientes esquemas son equivalentes:

$$\left\{\begin{array}{ll}i = 0          & \\
                   \textbf{while} & i<101:\\
                                  & \textrm{...}\\
                                  & i = i+1\end{array}\right.
                                     \quad\quad\leftrightarrow \quad\quad
  \left\{\begin{array}{ll}\textbf{for}   & i  \textbf{  in}\textrm{ range(101)}:\\
                                  & \textrm{...}\end{array}\right.$$

El comando $range$ ademite algunas variantes, dependiendo de la cantidad de parámetros que se le den:
* range(n): empieza en 0, va sumando de a 1 y el último valor considerado es n-1.
* range(m,n): empieza en m, va sumando de a 1 y el último valor considerado es n-1.
* range(m,n,k): empieza en m, va sumando de a k y el último valor considerado es n-1.

Por ejemplo: si queremos sumar todos los pares entre 10 y 70 inclusive:

```python
suma = 0
for i in range(10,71,2):
    suma = suma+i
print(suma)
```

### If

El comando **if** verfica una condición y ejecuta el código correspondiente sólo si esa condición se cumple. NO realiza ningún loop. Por ejemplo:

```python
x = 5
if x==5:
    print('x vale 5')
```

**if** se puede complementar con **else**:

```python
x = 4
if x==5:
    print('x vale 5')
else:
    print('x no vale 5')
print('terminé')
```

Más en general, se pueden poner varias condiciones con **elif**:

```python
x = 27
if x==5:
    print('x vale 5')
elif x==6:
    print('x vale 6')
else:
    print('x no vale ni 5 ni 6')
```

**while** es el más general de los comandos. En nuestro primer ejemplo vimos que se puede usar un **while** para reemplazar un **for**. Es fácil verificar que también se puede usar para reemplazar un **if**. **for** e **if** son comandos simplificados para escribir menos código y más claro, cuando es posible. 

### Más ejemplos:

**1)** Dada la lista de valores [1,5,7,3,9,8,17], queremos sumar todos los elementos de la lista hasta que la suma exceda el número 10. Aquí tenemos un ciclo con _dos_ condiciones: una condición dice que no nos tenemos que pasar del tamaño de la lista y otra condición dice que no nos tenemos que pasar de un cierto valor para la suma. Para esto, conviene usar un **while**:

```python
a    = [1,5,7,3,9,8,17]
i    = 0
suma = 0
while i<len(a) and suma<10:
    suma = suma + a[i]
    i = i+1
print(suma)
```

**2)** Dado el vector $(-1,3,5,-8,2,9,3,-7,4,-4)$, queremos calcular la suma de todos los elementos _positivos_. Aquí también hay dos condiciones, pero son de distinto tipo: el ciclo sólo se ve afectado por la condición de no pasarse del tamaño de la lista. Luego, para decidir si un valor se incorpora a la suma o no hay que ver si es positivo o negativo: pero esto no afecta el ciclo. Por lo tanto, el modo más simple de programar esto es combinar un **for** con un **if**:

```python
a    = np.array([-1,3,5,-8,2,9,3,-7,-2,4,-4])
suma = 0
for i in range(len(a)):
    if a[i]>0:
        suma = suma + a[i]
print(suma)
```

Podemos adaptar el programa anterior para que sume simultáneamente por un lado los positivos y por otro los negativos:


```python
a        = np.array([-1,3,5,-8,2,9,3,-7,-2,4,-4])
suma_pos = 0
suma_neg = 0
for i in range(len(a)):
    if a[i]>0:
        suma_pos = suma_pos + a[i]
    else:
        suma_neg = suma_neg + a[i]
print('Los positivos suman:',suma_pos)
print('Los negativos suman:',suma_neg)
```
